```python
#  Imports
import os
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
#  Set API Key
os.environ["GOOGLE_API_KEY"] = userdata.get("GEMINI_API_KEY")
```

```python
#  Create Gemini LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash",
    temperature=0.7,
    convert_system_message_to_human=True
)
```

```python
#  Prompt to get user's name
name_prompt = PromptTemplate.from_template("My name is {name}. Please remember it and greet me.")
```

```python
#  Prompt to ask if it remembers the name
followup_prompt = PromptTemplate.from_template("Do you remember my name?")
```

```python
#  Create Chains
name_chain = LLMChain(llm=llm, prompt=name_prompt)
followup_chain = LLMChain(llm=llm, prompt=followup_prompt)
```

```python
#  Run both chains
name = input("Enter your name: ")
greet = name_chain.invoke({"name": name})  # First prompt - gives name
print("\n  Greet:\n", greet["text"])

print("\n Asking if it remembers your name...")
response = followup_chain.invoke({})  # Second prompt - ask if it remembers

print("\n  Response:\n", response["text"])
```
