```python
#  Imports
import os
from google.colab import userdata
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.memory import ConversationBufferMemory
```

```python
#  Set API Key
os.environ["GOOGLE_API_KEY"] = userdata.get("GEMINI_API_KEY")
```

```python
#  Create Gemini LLM
llm = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash",
    temperature=0.7,
    convert_system_message_to_human=True
)
```

```python

#  Add memory - using input_key and output_key for better control
memory = ConversationBufferMemory(
    memory_key="chat_history",
    input_key="input",
    output_key="text",
    return_messages=True
)
```

```python
#  Create a unified prompt template that can handle different types of messages
unified_prompt = PromptTemplate(
    template="Chat History: {chat_history}\n\nHuman: {input}\nAI:",
    input_variables=["input"],
    memory_template=memory
)
```

```python
#  Create a single chain with memory
conversation_chain = LLMChain(
    llm=llm,
    prompt=unified_prompt,
    memory=memory,
    verbose=True,
    output_key="text"
)
```

```python
#  Run the conversation
name = input("Enter your name: ")

# First interaction: tell your name
greet = conversation_chain.invoke({"input": f"My name is {name}. Please remember it and greet me."})
print("\n Greet:\n", greet["text"])

print("\n Asking if it remembers your name...")
# Second interaction: ask if it remembers
response = conversation_chain.invoke({"input": "Do you remember my name?"})
print("\n Response:\n", response["text"])
```
